**1. Importing data**

```python

import pandas as pd
df=pd.read_csv('/content/movies.csv')
df.head(10)
```

```python
#Taking the percentage of null value for each column
null_percentages = (df.isnull().sum() / len(df)) * 100
print("Percentage of null value for each column")
print(null_percentages)
```

**2. Data cleaning**

Changing data type

```python
df['votes'] = df['votes'].astype(pd.Int64Dtype())
df['budget'] = df['budget'].astype(pd.Int64Dtype())
df['gross'] = df['gross'].astype(pd.Int64Dtype())
print(df[['votes', 'budget', 'gross']].dtypes)
```

Split data

```python
df['Date released'] = df['released'].str.split(' \(', n=1).str[0]
df['Country released'] = df['released'].str.extract(r'\((.*?)\)', expand=False)
```

```python
df.head(10)
```

Sorting by votes

```python
df.sort_values(by='votes', ascending=False, inplace=False)
```

Dropping dup;icates


```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

**3. Correlation**

Budget for film vs Gross Earning

```python
import matplotlib.pyplot as plt

# Drop rows with NaN values in 'budget' or 'gross' to avoid plotting issues
df_plot = df.dropna(subset=['budget', 'gross'])

plt.figure(figsize=(10, 6))
plt.scatter(df_plot['budget'], df_plot['gross'])
plt.xlabel('Budget for film')
plt.ylabel('Gross Earning')
plt.title('Budget for film vs. Gross Earning')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with NaN values in 'budget' or 'gross' to avoid plotting issues
df_plot = df.dropna(subset=['budget', 'gross'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget', y='gross', data=df_plot)
sns.regplot(x='budget', y='gross', data=df_plot, scatter=False, color='red') # Add regression line
plt.xlabel('Budget for film')
plt.ylabel('Gross Earning')
plt.title('Budget for film vs. Gross Earning (Seaborn)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the Kendall correlation matrix
correlation_matrix = df.corr(numeric_only=True, method='kendall')

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numeric Features (Kendall Method)')
plt.show()
```

Numerizing dataset

```python
df_numerize = df.copy()

for col_name in df_numerize.columns:
    if df_numerize[col_name].dtype == 'object':
        df_numerize[col_name] = df_numerize[col_name].astype('category')
        df_numerize[col_name] = df_numerize[col_name].cat.codes

print(df_numerize.head())
```

```python
df_numerize.head(10)
```

Correlation with numberized data

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix_numerize = df_numerize.corr(numeric_only=True, method='pearson')

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_numerize, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerized Features (Pearson Method)')
plt.show()
```

Unstack and sort Coirrelation

```python
correlation_pairs = correlation_matrix_numerize.unstack()

# Sort the correlation values in descending order
sorted_correlation_pairs = correlation_pairs.sort_values(ascending=False)

# Optional: Filter out self-correlations
sorted_unique_correlation_pairs = sorted_correlation_pairs[sorted_correlation_pairs != 1.0]

# Filter out the specific pair ('Date released', 'released') and its inverse
filtered_correlation = sorted_unique_correlation_pairs[
    ~sorted_unique_correlation_pairs.index.map(lambda x: frozenset(x) == frozenset({'Date released', 'released'}))
]

# Remove duplicate pairs (e.g., (A,B) and (B,A) are the same correlation)
filtered_correlation = filtered_correlation.loc[lambda x: x.index.map(frozenset).duplicated() == False]

print(filtered_correlation.head(20))
```
